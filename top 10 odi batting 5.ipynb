```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting'

page = requests.get(url)

data = BeautifulSoup(page.content)
data
```

```python
new = []

for i in data.find_all("div", class_="rankings-block__banner--name-large"):
    new.append(i.text)
    
new
```

```python
res = []

for i in data.find_all("td", class_="table-body__cell rankings-table__name name"):
    
    res.append(i.text)
    
res
```

```python
Name = []
for sub in res:
    Name.append(sub.replace("\n", ""))
    
Name
```

```python
Country =[]

for i in data.find_all("span", class_="table-body__logo-text"):
    Country.append(i.text)
    
Country
```

```python
Rating =[]

for i in data.find_all("td", class_="table-body__cell rating"):
    Rating.append(i.text)
    
Rating
```

```python
import pandas as pd

top =pd.DataFrame()

top['Name'] = Name
top['Country'] = Country
top['Rating'] = Rating

tops=top.head(10)
tops
```
