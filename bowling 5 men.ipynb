```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'

page =requests.get(url)

data =BeautifulSoup(page.content, 'lxml')
```

```python
Country =[]

for i in data.find_all("td", class_="table-body__cell rankings-table__name name"):
    
    Country.append(i.text)
    

```

```python
import pandas as pd
country = []

for i in Country:
    country.append(i.replace('\n', ""))
    
countr =pd.DataFrame(country)
count=countr.head(10)
count
```

```python
team =[]

for i in data.find_all("span", class_="table-body__logo-text"):
    
    team.append(i.text)

Team=pd.DataFrame(team)
team=Team.head(10)
team
```

```python
import pandas as pd
Ranking =[]

for i in data.find_all("td", class_="table-body__cell rating"):
    
    Ranking.append(i.text)
    
ranking=pd.DataFrame(Ranking).head(10)
ranking
```

```python
bowling =pd.DataFrame()

bowling['count']=count
bowling['team']=team
bowling['ranking']=ranking

bowling
```
