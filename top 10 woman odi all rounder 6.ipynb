```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder'

page =requests.get(url)

data =BeautifulSoup(page.content, 'lxml')
data
```

```python
name =[]

for i in data.find_all("td", class_="table-body__cell rankings-table__name name"):
    
    name.append(i.text)
    
name
```

```python
Name =[]

for i in name:
    Name.append(i.replace('\n', ''))
Name
```

```python
Team =[]

for i in data.find_all("span", class_="table-body__logo-text"):
    Team.append(i.text)
Team
```

```python
Rating = []

for i in data.find_all("td", class_="table-body__cell rating"):
    Rating.append(i.text)
    
Rating
```

```python
import pandas as pd

Top = pd.DataFrame()

Top['Name'] = Name
Top['Team'] = Team
Top['Rating'] = Rating
TOP=Top.head(10)
TOP
```
