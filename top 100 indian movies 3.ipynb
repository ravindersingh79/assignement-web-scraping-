```python
import requests as re
from bs4 import BeautifulSoup
```

```python
url ='https://www.imdb.com/india/top-rated-indian-movies/'
page =re.get(url)
data=BeautifulSoup(page.content, 'lxml')
print(data.prettify())
```

```python
Movies_Name =[]

for i in data.find_all("td", class_="titleColumn"):
    
    Movies_Name.append(i.text)
    
```

```python
Movies =[]
for i in Movies_Name:
    i =i.strip()
    Movies.append(i.replace("\n", ""))
Movies
```

```python
Rating =[]

for i in data.find_all("td", class_="ratingColumn imdbRating"):
    
    Rating.append(i.text)
    
```

```python
Rate =[]
for i in Rating:
    Rate.append(i.replace("\n", ""))
```

```python
import pandas as pd

top =pd.DataFrame()

top['Movies']=Movies
top['Rate']=Rate

Top=top.head(100)
Top
```
