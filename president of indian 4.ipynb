```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://presidentofindia.nic.in/former-presidents.htm'
```

```python
page = requests.get(url)
```

```python
data = BeautifulSoup(page.content)
data
```

```python
Name =[]

for i in data.find_all('h3'):
   
    Name.append(i.text)
    
Name
```

```python
Term =[]

for i in data.find_all('p'):
   
    Term.append(i.text)
    
term=list(Term)
terms=str(term)
terms1=terms.split(":")
```

```python

import pandas as pd

frame =pd.DataFrame(terms1)
frame
```

```python
frame2=frame.drop([0,2,4,6,19,20,8])
```
